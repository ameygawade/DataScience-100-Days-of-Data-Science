## 1. Descriptive Statistics

Descriptive statistics summarize and describe the main features of a dataset. They provide simple summaries about the sample and the measures, helping to understand the basic features of the data.

#### Key Concepts:

- Mean
    - The average value of a dataset.
- Median
    - The middle value when the data is sorted in ascending or descending order.
- Mode
    -  The most frequent value in the dataset.
- Variance
    - Measures how much the data points deviate from the mean.
- Standard Deviation
    - The square root of the variance, indicating how spread out the data is.

### Example:

Suppose we have a dataset representing the ages of a group of people: 

[23,25,29,30,31,35,40,44,50].

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
ages = np.array([23, 25, 29, 30, 31, 35, 40, 44, 50])

# Calculate descriptive statistics
mean_age = np.mean(ages)
median_age = np.median(ages)
mode_age = np.argmax(np.bincount(ages))  # For small discrete datasets
variance_age = np.var(ages)
std_dev_age = np.std(ages)

# Display results
print(f"Mean Age: {mean_age}")
print(f"Median Age: {median_age}")
print(f"Mode Age: {mode_age}")
print(f"Variance: {variance_age}")
print(f"Standard Deviation: {std_dev_age}")

# Plotting the distribution of ages
plt.figure(figsize=(8, 6))
plt.hist(ages, bins=5, color='skyblue', edgecolor='black')
plt.axvline(mean_age, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_age}')
plt.axvline(median_age, color='green', linestyle='dashed', linewidth=1, label=f'Median: {median_age}')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

## 2. Probability Theory

Probability is the branch of mathematics that deals with the likelihood of events occurring. In data science, probability helps us to model uncertainty and make predictions based on data.The measure of the likelihood of an event occurring, ranging from 0 (impossible) to 1 (certain).

### Probability

#### Example: Predicting Customer Purchases

Imagine you are working for an online retail company and want to predict whether a customer will purchase a product based on historical data.

Suppose you know that out of 10,000 website visits, 2,000 resulted in a purchase. The probability $$ 𝑃(purchase) $$ of any given customer making a purchase is:

```python
total_number_of_website_visit = 10000
total_no_of_purchase = 2000

prob_purchase = total_no_of_purchase/ total_number_of_website_visit

print(f"Probability of purchase: {prob_purchase:.2f}")
```

This means there is a 20% chance that a randomly selected customer will make a purchase.

### Joint Probability

Probability of both of the event happening concurrently noted as $$ P( A ∩ B) $$ Also, can be written as $$ P(AB) $$

#### Example : Customer adds products to their cart and also purchase it

```python
# Define the probabilities
P_A = 0.3  # Probability of adding to cart
P_B_given_A = 0.2  # Probability of purchasing given that they added to cart

# Calculate the joint probability P(A ∩ B)
P_A_and_B = P_A * P_B_given_A

# Display the result
print(f"Probability that a customer adds the product to their cart and completes the purchase (P(A ∩ B)): {P_A_and_B:.2f}")
```

###  Union Probability

A probaility of atleast one occuring out of the two events is union probability and noted $$ P( A ∪ B) $$

And can be calculated as $$ P( A ∪ B) =  P(A) + P(B) - P(A ∩ B) $$

Which means hen we talk about the union of two events A andB in probability, denoted as $$ 𝑃(𝐴 ∪ 𝐵) $$, we are considering the probability that at least one of the events occurs. This includes three scenarios:

- Event A occurs (but not B).
- Event B occurs (but not A).
- Both events A and B occur.

#### Example: Probability that a Customer Either Visits the Site, Clicks on an Ad, or Does Both

Let’s consider an example where:

- Event A: A customer visits a website.
- Event B: The customer clicks on an advertisement on that websit

```python
# Given probabilities
P_A = 0.6  # Probability of visiting the website
P_B = 0.3  # Probability of clicking on the advertisement
P_A_and_B = 0.2  # Probability that the customer visits the site and clicks on the ad

# Calculate P(A ∪ B) using the formula: P(A ∪ B) = P(A) + P(B) - P(A ∩ B)
P_A_or_B = P_A + P_B - P_A_and_B

# Display the result
print(f"Probability that the customer either visits the site, clicks on the ad, or does both (P(A ∪ B)): {P_A_or_B:.2f}")
```

### Conditional Probability: 

 The probability of an event occurring given that another event has already occurred.

#### Example: Predicting Customer Behavior - Probability that a Customer Buys a Product Given That They Clicked on a Product Ad

Suppose you are analyzing customer behavior for an online store, and you want to calculate the probability that a customer will buy a product given that they have added the product to their shopping cart.

Let's define the events:

Event A: The customer buys the product.
Event B: The customer adds the product to the cart.
Conditional Probability Formula
The conditional probability of 𝐴 given 𝐵 is calculated as:
$$ 𝑃(𝐴∣𝐵) = 𝑃(𝐴 ∩ 𝐵)/P(B) $$
​


Where:
$$ 𝑃(𝐴 ∩ 𝐵) $$ 
is the probability that both events 𝐴 and 𝐵 occur (i.e., the customer adds the product to the cart and then buys it).
$$ P(B) $$ 
is the probability that event occurs (i.e., the customer adds the product to the cart).

```python
# Given probabilities
P_B = 0.4  # Probability that a customer clicks on the ad
P_A_and_B = 0.25  # Probability that a customer clicks on the ad and purchases the product

# Calculate P(A|B) using the formula: P(A|B) = P(A ∩ B) / P(B)
P_A_given_B = P_A_and_B / P_B

# Display the result
print(f"Probability that a customer buys the product given they clicked on the ad (P(A|B)): {P_A_given_B:.2f}")
```

### Complement Probability 

$ P(A^c) $ represents the probability that event A does not occur. This is known as the complement of  A.

#### Example: Predicting customer churn

Lets assume, You work for a subscription-based service and want to analyze the probability that a customer will not churn (cancel their subscription) within a given period. 
- If $ A $ is the event that a customer churns, then the complement $ A^c $  represents the event that the customer does not churn, meaning they stay subscribed.

```python
# Probability that a customer will churn (event A)
P_A = 0.25

# Complement probability: Probability that a customer will not churn (event A^c)
P_A_complement = 1 - P_A

# Display the result
print(f"Probability that a customer will not churn (P(A^c)): {P_A_complement:.2f}")
```

### Intersection with complement probabiliy

$ P(A∩B^c)$ is the probability that event $ 𝐴$ occurs while event $𝐵$ does not occur. This is also known as the intersection of $𝐴$ and the complement of $𝐵$.

#### Example: Understanding customers buying behaviour - probability that a customer views the product but does not make a purchase.

Scenario:
- Event A: A customer views a product on an e-commerce website.
- Event B: The customer buys the product.

$ 𝑃 ( 𝐴 ∩ 𝐵^𝑐) $ would represent the probability that a customer views the product but does not make a purchase.

```python
# Example data
total_customers = 1000  # Total number of customers observed
viewed_product = 400  # Number of customers who viewed the product
purchased_product = 100  # Number of customers who viewed and purchased the product

# Calculate P(A), the probability of viewing the product
P_A = viewed_product / total_customers

# Calculate P(A ∩ B), the probability of viewing and purchasing the product
P_A_and_B = purchased_product / total_customers

# Calculate P(A ∩ B^c), the probability of viewing the product but not purchasing it
P_A_and_B_complement = P_A - P_A_and_B

# Display the result
print(f"Probability that a customer views the product but does not make a purchase ( P(A ∩ B^c)): {P_A_and_B_complement:.2f}")
```

### Conditional Probability with Complement

$ P(A∣B^c) $ is the probability that event $ 𝐴 $ occurs given that event $ 𝐵 $ does not occur.

#### Example: Probability that a Customer Views a Product Given They Did Not Purchase It.

we'll calculate the conditional probability $ 𝑃 ( 𝐴 ∣ 𝐵^𝑐) $, which represents the likelihood that a customer views a product given that they did not purchase it.

Definitions:
- Event $ A $: The customer views a product.
- Event $ 𝐵^𝑐 $: The customer does not purchase the product.

$ P(A∣B^c) $ = $P(A∩B^c)/P(B^c) $
​


```python
# Example data
total_customers = 1000  # Total number of customers observed
viewed_product = 400  # Number of customers who viewed the product
purchased_product = 100  # Number of customers who viewed and purchased the product

# Calculate P(B^c), the probability that a customer did not purchase the product
P_B_complement = 1 - (purchased_product / total_customers)

# Calculate P(A), the probability of viewing the product
P_A = viewed_product / total_customers

# Calculate P(A ∩ B^c), the probability of viewing the product but not purchasing it
P_A_and_B_complement = P_A - (purchased_product / total_customers)

# Calculate P(A|B^c), the conditional probability that a customer views the product given they did not purchase it
P_A_given_B_complement = P_A_and_B_complement / P_B_complement

# Display the result
print(f"Probability that a customer views the product given they did not purchase it (P(A|B^c)): {P_A_given_B_complement:.2f}")
```

### Bayes theorm

Bayes' Theorem is a mathematical formula used to update the probability of a hypothesis as more evidence or information becomes available. It is widely used in various fields, including statistics, machine learning, and artificial intelligence.

$P(A∣B)= P(B∣A)⋅P(A)/P(B)​$


#### Example: Email Spam Detection

In spam detection, we often need to calculate the probability that an email is spam given that it contains specific words, like "free."

Let's calculate the probability $$ 𝑃(spam ∣ "free") $$, which is the probability that an email is spam given that it contains the word "free."

Lets consider
- $ A $: The event that the email is spam.

- $ B $: The event that the email contains the word "free."

Thus, by using bayes theoram we can calculate 
$ P(spam∣"free") $ = $ P("free"∣spam)⋅P(spam)$ | $ P("free") $

​
 


```python
# Given probabilities
P_spam = 0.3  # Probability that any email is spam
P_free_given_spam = 0.8  # Probability that "free" appears in a spam email
P_free = 0.5  # Probability that "free" appears in any email

# Applying Bayes' Theorem to calculate P(spam | "free")
P_spam_given_free = (P_free_given_spam * P_spam) / P_free

# Display the result
print(f"Probability that an email is spam given it contains 'free' (P(spam | 'free')): {P_spam_given_free:.2f}")
```

## 3. Distribution 

Probability distributions describe how the values of a random variable are distributed. The most common distributions are the normal, binomial, and Poisson distributions.

### Normal Distribution

The normal distribution, also known as the Gaussian distribution, is a continuous probability distribution that is symmetric around its mean. Most of the data points are concentrated around the mean, with the probability tapering off as you move away from the mean.

Characteristics:

- It has a bell-shaped curve.
- The mean, median, and mode are all equal.
- It is fully described by two parameters: the mean $ (μ) $ and the standard deviation $ (σ) $.

#### Example : Application in Quality Control 

Suppose a factory produces light bulbs, and the lifespan of these bulbs follows a normal distribution with a mean of 1,000 hours and a standard deviation of 100 hours. By analyzing the distribution, the factory can determine the proportion of bulbs that are likely to fail before 900 hours or last beyond 1,100 hours, which helps in maintaining quality standards.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Mean and standard deviation
mu, sigma = 1000, 100

# Generate data
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
y = norm.pdf(x, mu, sigma)

# Plot
plt.plot(x, y)
plt.title('Normal Distribution')
plt.xlabel('Lifespan (hours)')
plt.ylabel('Probability Density')
plt.show()
```

### Binomial Distribution

The binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials, where each trial has two possible outcomes (like success/failure or yes/no).

Characteristics:

- It is defined by two parameters: the number of trials $(n)$ and the probability of success in each trial $(p)$.
- The mean of the distribution is $ 𝑛𝑝 $ and the variance is $ 𝑛𝑝 (1−𝑝) $.

#### Example: Application in A/B testing

Suppose you are running an A/B test to determine if a new website design leads to more user sign-ups compared to the old design. The binomial distribution can model the number of sign-ups (successes) in a sample of users who saw the new design, where each user's response is a Bernoulli trial (sign-up or not).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Number of trials and probability of success
n, p = 10, 0.5

# Generate data
x = np.arange(0, n+1)
y = binom.pmf(x, n, p)

# Plot
plt.bar(x, y)
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
```

### Poisson Distribution

The Poisson distribution is a discrete probability distribution that describes the number of events occurring within a fixed interval of time or space, given that these events occur with a known constant mean rate and independently of the time since the last event.

Characteristics:

- It is defined by the parameter $ λ $ (the average number of events in the interval).
- The mean and variance of the distribution are both equal to $ λ $.

#### Example: Application in call center analytics

Suppose a call center receives an average of 10 calls per hour. The Poisson distribution can model the number of calls received in an hour, helping the call center manage resources effectively by predicting periods of high or low call volumes.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Average number of events (calls per hour)
lambda_ = 10

# Generate data
x = np.arange(0, 30)
y = poisson.pmf(x, lambda_)

# Plot
plt.bar(x, y)
plt.title('Poisson Distribution')
plt.xlabel('Number of Calls')
plt.ylabel('Probability')
plt.show()
```
