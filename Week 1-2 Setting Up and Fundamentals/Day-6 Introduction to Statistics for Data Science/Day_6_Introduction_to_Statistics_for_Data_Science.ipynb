## 1. Descriptive Statistics

Descriptive statistics summarize and describe the main features of a dataset. They provide simple summaries about the sample and the measures, helping to understand the basic features of the data.

#### Key Concepts:

- Mean
    - The average value of a dataset.
- Median
    - The middle value when the data is sorted in ascending or descending order.
- Mode
    -  The most frequent value in the dataset.
- Variance
    - Measures how much the data points deviate from the mean.
- Standard Deviation
    - The square root of the variance, indicating how spread out the data is.

### Example:

Suppose we have a dataset representing the ages of a group of people: 

[23,25,29,30,31,35,40,44,50].

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
ages = np.array([23, 25, 29, 30, 31, 35, 40, 44, 50])

# Calculate descriptive statistics
mean_age = np.mean(ages)
median_age = np.median(ages)
mode_age = np.argmax(np.bincount(ages))  # For small discrete datasets
variance_age = np.var(ages)
std_dev_age = np.std(ages)

# Display results
print(f"Mean Age: {mean_age}")
print(f"Median Age: {median_age}")
print(f"Mode Age: {mode_age}")
print(f"Variance: {variance_age}")
print(f"Standard Deviation: {std_dev_age}")

# Plotting the distribution of ages
plt.figure(figsize=(8, 6))
plt.hist(ages, bins=5, color='skyblue', edgecolor='black')
plt.axvline(mean_age, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_age}')
plt.axvline(median_age, color='green', linestyle='dashed', linewidth=1, label=f'Median: {median_age}')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

## 2. Probability Theory

Probability is the branch of mathematics that deals with the likelihood of events occurring. In data science, probability helps us to model uncertainty and make predictions based on data.The measure of the likelihood of an event occurring, ranging from 0 (impossible) to 1 (certain).

### Probability

#### Example:

```python
# Probability of drawing an Ace
prob_ace = 4 / 52  # There are 4 Aces in a deck of 52 cards
print(f"Probability of drawing an Ace: {prob_ace:.2f}")
```
