#### Example:

##### Predicting House Prices:
In the real estate industry, a common use of linear equations is to predict house prices based on certain features. For example, a linear equation could be used to model the relationship between the price of a house (y) and its size in square feet (x).
- Equation: y=mx+c

Suppose a dataset shows that for every additional square foot, the price increases by $150, and a 0-square-foot house theoretically costs $50,000 (this is the y-intercept, although not practical).

- Linear Equation: Price=150Ã—Size+50000

```python
import numpy as np
import matplotlib.pyplot as plt

# Linear equation parameters
m = 150  # Slope (price per square foot)
c = 50000  # Intercept (base price)

# Function to calculate house price based on size
def predict_price(size):
    return m * size + c

# Example sizes (square feet)
house_sizes = np.array([1000, 1500, 2000, 2500, 3000])

# Predicting prices for the given house sizes
predicted_prices = predict_price(house_sizes)

# Display the results
for size, price in zip(house_sizes, predicted_prices):
    print(f"House Size: {size} sq. ft. -> Predicted Price: ${price}")

# Plotting the linear relationship
plt.figure(figsize=(8, 6))
plt.plot(house_sizes, predicted_prices, color='blue', marker='o', linestyle='-', label='Predicted Price')
plt.title('House Price Prediction based on Size')
plt.xlabel('Size (square feet)')
plt.ylabel('Price ($)')
plt.grid(True)
plt.legend()
plt.show()
```

### Quadratic Equations 

The quadratic equation is:

$$ ax^2 + bx + c = 0 $$

The solutions to the quadratic equation are given by:

$$ x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a} $$


#### Example:

##### Predicting Product Sales Based on Price

In retail, the relationship between product price and sales volume can sometimes be quadratic. If a product is priced too low, it might sell a lot, but with little profit margin. If itâ€™s priced too high, sales might drop. The optimal price point often lies somewhere in the middle, forming a parabolic relationship.

- Equation:
Letâ€™s say the relationship between price (x) and sales volume (y) is:
$$ y = ax^2+bx+c $$

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import Polynomial
from sklearn.metrics import mean_squared_error

# Generate sample data for price (x) and sales (y)
# Hypothetical data: As the price increases, sales first increase, then decrease
prices = np.array([5, 10, 15, 20, 25, 30])
sales = np.array([50, 80, 120, 100, 60, 30])

# Fit a quadratic model (2nd degree polynomial)
p = Polynomial.fit(prices, sales, 2)

# Generate a range of prices to predict sales
price_range = np.linspace(0, 35, 100)
predicted_sales = p(price_range)

# Plot the data and the quadratic model
plt.scatter(prices, sales, color='red', label='Observed Sales')
plt.plot(price_range, predicted_sales, color='blue', label='Quadratic Fit')
plt.xlabel('Price')
plt.ylabel('Sales')
plt.title('Predicting Product Sales Based on Price')
plt.legend()
plt.grid(True)
plt.show()
```

## 2.Calculus

### Derivatives

The derivative of a function measures the rate at which the function's value changes as the input changes. It is a fundamental concept in calculus and is crucial for understanding optimization problems, especially in machine learning techniques like gradient descent.

```python
import numpy as np
import matplotlib.pyplot as plt

# Hypothetical data: square footage and corresponding house prices
square_footage = np.array([1000, 1500, 2000, 2500, 3000])
prices = np.array([200000, 250000, 300000, 350000, 400000])

# Standardize the input feature
square_footage = (square_footage - np.mean(square_footage)) / np.std(square_footage)

# Initialize model parameters (e.g., slope and intercept for a simple linear model)
theta_0 = 0  # intercept
theta_1 = 0  # slope for square footage

# Learning rate
alpha = 0.01  

# Number of iterations
iterations = 200

# Lists to store the values for plotting
theta_0_history = []
theta_1_history = []
loss_history = []

# Gradient Descent to minimize the loss
for _ in range(iterations):
    # Predict prices using the current model parameters
    predicted_prices = theta_0 + theta_1 * square_footage
    
    # Calculate the loss (Mean Squared Error)
    loss = np.mean((predicted_prices - prices) ** 2)
    
    # Calculate gradients (derivatives of the loss function)
    gradient_theta_0 = np.mean(predicted_prices - prices)
    gradient_theta_1 = np.mean((predicted_prices - prices) * square_footage)
    
    # Update parameters
    theta_0 -= alpha * gradient_theta_0
    theta_1 -= alpha * gradient_theta_1
    
    # Store the values for plotting
    theta_0_history.append(theta_0)
    theta_1_history.append(theta_1)
    loss_history.append(loss)

# Plotting the results
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# Plot the loss over iterations
axs[0].plot(range(iterations), loss_history, label='Loss', color='red')
axs[0].set_xlabel('Iteration')
axs[0].set_ylabel('Loss')
axs[0].set_title('Loss Over Iterations')
axs[0].grid(True)

# Plot the predicted price curve after training
predicted_prices_final = theta_0 + theta_1 * square_footage
axs[1].scatter(square_footage, prices, color='blue', label='Actual Prices')
axs[1].plot(square_footage, predicted_prices_final, color='green', label='Fitted Line')
axs[1].set_xlabel('Square Footage (Standardized)')
axs[1].set_ylabel('Price')
axs[1].set_title('Predicted Prices vs. Actual Prices')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

### Integrals

 The integral of a function represents the area under the curve of that function on a graph. It is a fundamental concept in calculus used to calculate quantities like area, volume, displacement, and more. Integrals help in determining the accumulated quantity, which is particularly useful in fields like physics, engineering, and economics.

The integral of a function $$ f(x)  $$ is written as:

$$
\int f(x) \, dx
$$

This is read as "the integral of \( f(x) \) with respect to \( x \)".


#### Why Do We Care About Integrals in Data Science?

One practical application of integrals in data science is in evaluating the performance of models, particularly in classification tasks.



#### Example: Area Under the ROC Curve (AUC)



When youâ€™re building a model that classifies things into categories (like spam vs. not spam emails), you want to know how good your model is at distinguishing between the categories. One way to do this is by looking at a graph called the Receiver Operating Characteristic (ROC) curve.

ROC Curve: This graph shows the trade-off between two things: the true positive rate (how many correct positive predictions your model makes) and the false positive rate (how many incorrect positive predictions your model makes).

Area Under the Curve (AUC): The AUC is a number that tells you how good your model is overall. AUC is calculated using an integralâ€”itâ€™s the area under the ROC curve.

```python
import numpy as np
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt

# Hypothetical data: True labels and predicted probabilities
y_true = np.array([0, 0, 1, 1])
y_scores = np.array([0.1, 0.4, 0.35, 0.8])

# Calculate the ROC curve
fpr, tpr, thresholds = roc_curve(y_true, y_scores)

# Calculate the AUC
auc = roc_auc_score(y_true, y_scores)
print(f'AUC: {auc:.2f}')

# Plotting the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line for random classifier
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

## 3.Linear algebra

### Vectors

A vector is an ordered list of numbers. Vectors can represent a point in space, where each number corresponds to a coordinate along a different axis.

A vector $$ \mathbf{v} $$  with components $$ v_1, v_2, v_3, \dots $$ can be written as:

$$
\mathbf{v} = \begin{pmatrix}
v_1 \\
v_2 \\
v_3 \\
\vdots \\
v_n
\end{pmatrix}

$$

#### Example:

In data science, each row in a dataset can be thought of as a vector. For instance, if you're analyzing customer data, a vector could represent the features of a single customer, such as age, income, and number of purchases.

###  Matrices

A matrix is a two-dimensional array of numbers, with rows and columns. Matrices are used to represent datasets, transformations, and systems of linear equations.

A matrix $$ A $$ with elements $$ a_{ij} $$ can be written as:

$$
A = \begin{pmatrix}
a_{11} & a_{12} & \dots  & a_{1n} \\
a_{21} & a_{22} & \dots  & a_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
a_{m1} & a_{m2} & \dots  & a_{mn}
\end{pmatrix}
$$


### Example

In a recommendation system, a matrix can represent the user-item interactions, where each row is a user, each column is an item, and each element is the user's rating or interaction with the item.
