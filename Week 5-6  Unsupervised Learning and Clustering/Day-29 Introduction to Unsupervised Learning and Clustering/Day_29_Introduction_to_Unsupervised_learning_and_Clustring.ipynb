# Day 29: Introduction to Unsupervised Learning and Clustering

Welcome to Day 29! Today, we transition from supervised to unsupervised learning. Unlike supervised learning which we learned in week 1 and 2, where we train models on labeled data to make predictions, unsupervised learning focuses on finding patterns and structures in unlabeled data.

This is a powerful technique used for tasks like customer segmentation, anomaly detection, and data compression. The primary goal of unsupervised learning is to gain insights from data that doesn't have a specific target variable to predict.

## What is Unsupervised Learning?

In supervised learning, you have both input features (X) and a target variable (y). The goal is to learn a function that maps X to y. 

    Think of it as a teacher giving you a quiz with an answer key.

In unsupervised learning, you only have the input features (X). There is no target variable to guide the model. The goal is to find hidden patterns, group similar data points, or reduce the dimensionality of the data. 

    Think of it as a student being given a set of blocks and being asked to sort them without any instructions.

    Analogy: Sorting Your LEGOs
- Supervised Learning: You're given a box of LEGOs with a blueprint for a car. Your task is to use the right pieces to build the car. The blueprint is the "label" or "target."

- Unsupervised Learning: You're given a large box of mixed LEGOs. Your task is to sort them by color, size, or shape. There is no blueprint; you have to discover the patterns on your own.

## Introduction to Clustering

Clustering is a type of unsupervised learning that groups data points into clusters so that data points in the same group are more similar to each other than to those in other groups.

It's a foundational technique used in many real-world applications, such as:

- Customer Segmentation: Grouping customers with similar purchasing habits.

- Image Segmentation: Dividing an image into different regions or objects.

- Anamoly Detection: Identifying unusual data points that don't fit into any cluster.

    Example: Generating Clusters with scikit-learn
We can use scikit-learn's `make_blobs` function to create a synthetic dataset that is perfect for demonstrating clustering. 
It generates **isotropic Gaussian blobs** of data, meaning the clusters are roughly circular and of similar density.

The X and y generated by `make_blobs` are our features and labels respectively. However, in an unsupervised setting, we would only use X for our clustering algorithm. We would use the labels y only to verify how well our algorithm performed.

### Data Generation

The make_blobs function helps us simulate this kind of data. We can tell it to create 1,000 data points (n_samples=1000) and to group them into three distinct clusters (n_clusters=3).

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=1000, n_features=2, centers=3, random_state=42)
```

### Visualising data witj respect to Unsupervised and Supervised learning

```python
import matplotlib.pyplot as plt
# --- Supervised Learning: Plotting with Known Labels ---
# In supervised learning, we have the labels 'y' available.
# We can use these labels to color the data points, showing the true clusters.
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', marker='o', edgecolor='k', s=50)
plt.title('Supervised Learning: Labeled Data (X and y)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# --- Unsupervised Learning: Plotting without Labels ---
# In unsupervised learning, the labels 'y' are not available for the algorithm.
# We plot all the data points in a single color to reflect this.
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], color='gray', marker='o', edgecolor='k', s=50)
plt.title('Unsupervised Learning: Unlabeled Data (X only)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Display the plots
plt.tight_layout()
plt.show()
```

### Proving working of Unsupervised Clustring with K-means clustring algorithm

```python
from sklearn.cluster import KMeans

# We don't use 'y' here because it's an unsupervised task
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
predicted_clusters = kmeans.predict(X)
```

```python
# Get the coordinates of the cluster centers
centroids = kmeans.cluster_centers_

# 3. Visualize the clusters
plt.figure(figsize=(8, 6))

# Plot the data points, colored by their predicted cluster
plt.scatter(X[:, 0], X[:, 1], c=predicted_clusters, cmap='viridis', marker='o', edgecolor='k', s=50)

# Plot the cluster centers (centroids)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='red', edgecolor='black', label='Centroids')

# Add titles and labels for clarity
plt.title('K-Means Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
